```python
# x, y = mnist.data, mnist.target
x, y = mnist['data'], mnist['target']
x.shape, y.shape
```

```python
import matplotlib.pyplot as plt
n_samples = 8

fig, axes = plt.subplots(1, n_samples, figsize=(10, 3))
for i in range(n_samples):
    # each row of X is a flattened 28Ã—28 image
    img = x[i].reshape(28, 28)
    axes[i].imshow(img, cmap='gray')
    axes[i].set_title(f"Label: {y[i]}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def plot_distribution(y_full, y_train, y_test):
    labels = np.unique(y_full)
    full_counts = [np.sum(y_full == label) for label in labels]
    train_counts = [np.sum(y_train == label) for label in labels]
    test_counts = [np.sum(y_test == label) for label in labels]

    x = np.arange(len(labels))
    width = 0.3

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(x - width, full_counts, width, label='Full Dataset')
    ax.bar(x, train_counts, width, label='Train Set')
    ax.bar(x + width, test_counts, width, label='Test Set')

    ax.set_xlabel('Digit Label')
    ax.set_ylabel('Count')
    ax.set_title('Label Distribution in Full, Train, and Test Sets')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    plt.show()
```

```python
# train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python

plot_distribution(y, y_train, y_test)
```
